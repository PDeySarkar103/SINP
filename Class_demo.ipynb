# Writing a class for OOP approach in python

## Writing a simple class for complex numbers

```python
class Complex:
  def __new__(cls,*args):
    # If user passes multiple complex numbers as tuples, this returns multiple # complex objects
    if len(args)>1 and all(isinstance(p,(tuple,list)) for p in args):
      result=[]
      for r,i in args:
        obj=super().__new__(cls)
        obj.real=r
        obj.imag=i
        result.append(obj)
      return tuple(result)
    #Otherwise return a single object
    return super().__new__(cls)
  def __init__(self,real,imag):
    #Initializes the real and imaginary parts
    self.real=real
    self.imag=imag
  def __str__(self):
    sign="+" if self.imag>=0 else "-"
    return f"{self.real}{sign}{abs(self.imag)}i"
  def __add__(self,other):
    return Complex(self.real+other.real,self.imag+other.imag)
  def __sub__(self,other):
    return Complex(self.real-other.real,self.imag-other.imag)
  def __mul__(self,other):
    r=self.real*other.real-self.imag*other.imag
    i=self.real*other.imag+self.imag*other.real
    return Complex(r,i)
  def conjugate(self):
    return Complex(self.real,-self.imag)
  def norm(self):
    return (self.real**2+self.imag**2)**0.5
```

```python
a,b= Complex((2,3),(4,5))
print("a=" ,Complex.__str__(a))
print("b=",Complex.__str__(b))
print("a+b=",Complex.__add__(a,b))
print("a-b=",Complex.__sub__(a,b))
print("a*b=",Complex.__mul__(a,b))
print("|a|=",Complex.norm(a))
print("|b|=",Complex.norm(b))
print("conjugate of a=",Complex.conjugate(a))
print("conjugate of b=",Complex.conjugate(b))
```
